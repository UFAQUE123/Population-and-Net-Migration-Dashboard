```python
# import libraries
import wbdata
import pandas as pd
```

# World Bank Indicator Codes
- These are official World Bank API indicators:
-  **SP.POP.TOTL**	Total population of a country
-  **SM.POP.NETM**	Net migration (immigrants − emigrants)
-  This dictionary is typically used to fetch data from the World Bank API

```python
# Load the data
indicator = {
    'SP.POP.TOTL': 'total_population',  # Population
    'SM.POP.NETM': 'net_migration'     # Net Migration
}
data = wbdata.get_dataframe(indicator)
data.reset_index(inplace=True)
data
```

```python
data.info()
```

```python
data.rename(columns={'country': 'Country', 'date': 'Year'}, inplace=True)
data['Year'] = pd.to_numeric(data['Year'])
data = data[(data['Year'] >= 1960) & (data['Year'] <= 2025)]
data
```

# Net Migration Values
- Positive → more people entering the country(immigrants)
- Negative → more people leaving the country(emigrants)
